```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import os

# Fetch the Mushroom dataset from UCI (ID = 73)
mushroom = fetch_ucirepo(id=73)

# Extract features (X) and target labels (y)
X = mushroom.data.features
y = mushroom.data.targets

# Display first few rows for verification
print("Features:")
display(X.head())

print("Target:")
display(y.head())
```

```python
# Define path
save_path = "C:\Repos\ml_classification_aanchal\data"

# Save features and target as CSV
X.to_csv(os.path.join(save_path, "mushroom_features.csv"), index=False)
y.to_csv(os.path.join(save_path, "mushroom_target.csv"), index=False)

print("✅ Files saved successfully in the 'data' folder.")
```
